***Resumo:*** Este trabalho tem como base a manipulação de listas, tendo como principal objetivo o ordenamento das mesmas com recurso ao algoritmo Bubble Sort. Neste algoritmo verifica-se a iteração de toda a lista e, simultaneamente, a comparação e ordenação de pares de valores, até que a lista esteja totalmente ordenada.<br>
Inicialmente é apresentado ao utilizador um menu que permite aceder às várias funções desenvolvidas, nomeadamente, a criação de uma lista manualmente ou, aleatoriamente. Além disso, pode-se escrever a lista corrente e , ainda, determinar o valor do maior elemento de uma dada lista.<br> 
Por último, é possível ordenar os elementos da lista por ondem crescente, através do algoritmo Bubble Sort, no qual, a posição atual é comparada com a posição seguinte. No caso de a posição atual ser superior à posição seguinte é efetuada a troca entre os elementos destas duas posição caso contrário, não se realiza troca alguma e passa-se para o par de elementos seguintes, a serem comparados. Para isso, inicializámos com a variável de controlo "trocas" a True A lista será iterada o número de vezes necessárias até que se encontre completamente ordenada, neste momento o algoritmo encontra o seu fim.   
</p>

```python
import random

def menu():
    print (""" Bem vindo!!
    (1) Gerar uma lista manualmente;
    (2) Gerar uma lista de valores aleatórios;
    (3) Escrever a lista corrente no stdout;
    (4) Maior elemento da lista;
    (5) Ordenar com o bubbleSort;
    (0) Sair""")
    
def criarListaM():
    lista = []
    n = int(input ("Introduza o número de elementos da lista: "))
    i = 1
    while len(lista)<n:
        elem = input ("Introduza o elemento " + str(i) + ":")
        lista.append(elem)
        i = i+1
    return lista



def criarListaA():
    lista = []
    from random import randrange 
    n = int(input("Introduza o número de elementos da lista: "))
    i = 1
    while len(lista)<n:
        elem = randrange (0,101)
        lista.append(elem)
        i = i+1
    return lista




def maiornumero(lista):
    maior = lista[0]
    for i in range(len(lista)):
        if lista[i]>maior:
            maior = lista[i]
    print("O maior elemento da lista é o número " + str(maior) + ".")
    return maior 





def bubbleSort (lista):
    trocas = True
    while trocas:
        trocas = False
        for i in range (len(lista)-1):
            if lista[i]>lista[i+1]:
                lista[i], lista[i+1]=lista[i+1], lista[i]
                trocas = True
    print("Esta é a lista ordenada: " + str(lista))
    return lista 



    
opcao = "1"
lista = []
while opcao != "0":
    menu()
    opcao = input("Introduza uma opção: ")
    if opcao == "1":
        lista = criarListaM()
        print(lista)
    elif opcao == "2":
        lista = criarListaA()
        print(lista)
    elif opcao == "3":
        print(lista)
    elif opcao == "4":
        maiornumero(lista)
    elif opcao == "5":
        bubbleSort(lista)
    elif opcao == "0":
        print ("Até à próxima!!")
    else:
        print ("Opção inválida!")
```
