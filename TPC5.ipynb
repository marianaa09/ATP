***Resumo:*** Este trabalho tem como principal objetivo o desenvolvimento de uma aplicação que permita a realização de várias operações com polinómios.<br>
Primeiramente, é mostrado ao utilizador um menu no qual são apresentadas as diferentes opções que a aplicação permite efetuar. Entre as várias funções está presente a criação de um polinómio, a sua posterior visulização, a determinação do seu grau, bem como o cálculo do valor do polinómio num ponto específico escolhido pelo utilizador. Além disso, é possível determinar o valor da sua derivada e simplificar o polinómio criado inicialmente. Por fim, é permitida a criação de uma tabela que apresenta o cálculo do valor do polinómio em vários pontos, que variam consoante o número de linhas especificado pelo utilizador, incluindo o número 0.<br>     


```python
def criarTermo(coef, exp):
    return (coef, exp)

def criarPolinomio():
    return []

def inserirTermo(p,t):
    p.append(t)
    
    
def verTermo(t):
    c, e = t
    if e == 0:
        return str(c)
    elif c == 1:
        return "x^" + str(e)
    else:
        return str(c) + "x^" + str(e)
        

        
def verPolinomio(p):
    politexto = ""
    if len(p) != 0:
        politexto = verTermo(p[0])
        for i in range(1,len(p)):
            if p[i][0] >= 0:
                politexto = politexto + "+" + verTermo(p[i])
            else:
                politexto = politexto + verTermo(p[i])
    return politexto



def grauPol(p):
    grau = p[0][1]
    for i in range(1,len(p)):
        if p[i][1] > grau:
            grau = p[i][1]
    
    return grau




def calcPol(p,x):
    valor = 0 
    for (c,e) in p:
        valor = valor + c * x**e
    return valor 



def derivadaPol (p):
    derivada = []
    for (c, e) in p:
        if e != 0:
            derivada.append((c*e, e-1))
    return derivada 



def tabela(p, linhas):
    print("x" + " :: " + verPolinomio(p))
    for l in range(0, linhas):
        print(str(l) + " :: " + str(calcPol(p, l)))
        

        
def simplificarPol(p):
    
    fst,last = 0, len(p)
    i = 0
    while(i<last) :
        somaCoefs = p[i][0]
        j = i
        while(j<last) :
            if i != j and p[i][1] == p[j][1] :
                somaCoefs += p[j][0]
                del(p[j])
                last = last - 1
            else :
                j = j+1
        p[i] = (somaCoefs,p[i][1])
        i = i+1
    return p

def menu():
    print("""Bem vindo!! Seleciona uma opção: 
    (1) Criar um polinómio
    (2) Listar o polinómio
    (3) Grau do polinómio
    (4) Calcular o polinómio num ponto
    (5) Calcular a derivada do polinómio 
    (6) Simplificar o polinómio
    (7) Tabela
    (0) Sair """)

opcao = "1"
while opcao != "0":
    menu()
    opcao = input("Introduza uma opção: ")
    if opcao == "1":
        print("----------Criar Polinómio----------")
        p = criarPolinomio()
        n = int(input("Quantos termos terá o polinómio?"))
        for i in range(n):
            coef = int(input("Introduza o coeficiente do termo " + str(i+1) + ":"))
            exp = int(input("Introduza o expoente do termo " + str(i+1) + ":"))
            inserirTermo(p,(coef,exp))
        criarPolinomio()
    elif len(p) == 0:
            print("O polinómio ainda não foi criado!")
            
    elif opcao == "2":
        print("----------Listar Polinómio----------")
        print(verPolinomio(p))
    
    elif opcao == "3":
        print("----------Grau do Polinómio----------")
        print("O grau do polinómio é: ", grauPol(p))
    elif opcao == "4":
        print("----------Calcular Polinómio num Ponto----------")
        x = int(input("Insira o valor para o qual quer calcular o polinómio: "))
        print("O valor do polinómio em " + str(x) + "é: ", calcPol(p,x))
    elif opcao == "5":
        print ("----------Calcular a Derivada do Polinómio----------")
        print("A derivada do polinómio é: ", verPolinomio(derivadaPol(p)))
    elif opcao == "6":
        print("----------Simplificar o Polinómio----------")
        print("Polinómio simplificado: ", verPolinomio(simplificarPol(p)))
    elif opcao == "7":
        print("----------Tabela----------")
        nlinhas = int(input("Número de linhas da tabela: "))
        print("x :: Valor")
        tabela(p,nlinhas)
    elif opcao == "0":
        print("Até à próxima!!")
    else:
        print("Opção inválida!!")

```
