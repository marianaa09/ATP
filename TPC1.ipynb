```python
# Jogo dos 21 fósforos


import random

def printMenu():
    menu = '*****************MENU******************\n\n1 -> Jogador vs Pc\n2 -> Pc vs Jogador\n0 -> Sair\n\n'
    menu = menu + '***************************************'
    return menu

def acceptOption():
    option = input('Insira uma opção válida: ')
    return option

def ultimo(fosforos):
    if fosforos == 1:
        return True
    
def novoJogo():
    
    validAnswer = False
    while not validAnswer :
        answer = input('Pretende iniciar um novo jogo? (Y/N)')
        if(answer == 'Y'):
            validAnswer = True
            return True
        elif answer == 'N':
            validAnswer = True
            return False
    

def jogadorVsPc(fosforos):
    
    while fosforos != 0 :
        ''' Mostramos para o jogador a quantidade de fósforos que existe no momento '''
        print('Neste momento existem ' + str(fosforos) + ' fósforos.')
        
        ''' Jogada feita por parte do jogador '''
        while True:
            fosfJogador = int(input('[JOGADOR] Insira o nº de fósforos a retirar: '))
            if fosfJogador<=4 and fosfJogador<=fosforos:
                break
            else:
                print('O valor inserido é inválido...')
        fosforos = fosforos - fosfJogador
        print('[JOGADOR] Foram removidos ' + str(fosfJogador) + ' fósforos pelo jogador.')
        
        if ultimo(fosforos):
            print('FIM DO JOGO! VENCEDOR: JOGADOR')
            break
            
        ''' Jogada feita pelo PC '''
        fosfPC = 5 - fosfJogador
        fosforos = fosforos - fosfPC
        print('[PC] Foram removidos ' + str(fosfPC) + ' fósforos pelo PC.')
        
        if ultimo(fosforos):
            print('FIM DO JOGO! VENCEDOR: PC')
            break
            
            
def pcVsJogador(fosforos):
    
    while fosforos != 0 :
        ''' Mostramos para o jogador a quantidade de fósforos que existe no momento '''
        print('Neste momento existem ' + str(fosforos) + ' fósforos.')
        
        ''' Jogada feita pelo PC '''
        if(fosforos <= 5) :
            fosfPC = fosforos-1
        else :
            fosfPC = random.randint(1,4)
        fosforos = fosforos - fosfPC
        print('[PC] Foram removidos ' + str(fosfPC) + ' fósforos pelo PC.')
        
        if ultimo(fosforos):
            print('FIM DO JOGo! VENCEDOR: PC')
            break
            
        ''' Jogada feita por parte do jogador '''
        while True:
            fosfJogador = int(input('[JOGADOR] Insira o nº de fósforos a retirar: '))
            if fosfJogador<=4 and fosfJogador<=fosforos:
                break
            else:
                print('O valor inserido é inválido...')
        fosforos = fosforos - fosfJogador
        print('[JOGADOR] Foram removidos ' + str(fosfJogador) + ' fósforos pelo jogador.')
        
        if ultimo(fosforos):
            print('FIM DO JOGO! VENCEDOR: JOGADOR')
            break
        
        
        
        

def main():
    fosforos = 21
    print(printMenu())
    option = int(acceptOption())
    while option != 0:
        fosforos = 21
        
        if option == 1:
            jogadorVsPc(fosforos)
            while novoJogo() :
                jogadorVsPc(21)
        elif option == 2:
            pcVsJogador(fosforos)
            while novoJogo() :
                pcVsJogador(21)
        print(printMenu())
        option = int(acceptOption())
    print('Até à próxima!!!!')
    
    

    

    
    
```

```python
if __name__ == "__main__":
    main()    
    
```
