***Resumo:*** Neste trabalho, o principal objetivo foi a criação de um menu de uma aplicação que permita a realização de várias operações com frações.<br>
As funções desenvolvidas possibilitam ao utilizador criar frações, visualizá-las e realizar as operações básicas entre duas frações. Além disso, o utilizador pode gerar uma lista com um número, pré determinado, de frações aleatórias. São concedidas, ainda, as opções que permitem elaborar a soma e multiplicação das frações que constituem a lista e determinar o menor e o maior elemento presente.



```python
import random 


def criarFracao(numerador, denominador):
    return(numerador,denominador)


def verFracao(f):
    return (str(f[0]) + "/" + str(f[1]))

def verListFrac(lista):
    c = []
    for frac in lista:
        c.append(verFracao(frac))
    return c


def mdc(a,b):
    if a%b==0:
        return b
    elif b%a==0:
        return a
    else:
        if a>b:
            return mdc(b, a%b)
        else:
            return mdc(a, b%a) 

def simplificarFracao(f):
    n,d = f
    m = mdc(n,d)
    return (int(n/m), int(d/m))


def somarFrac(f1,f2):
    n1,d1 = f1
    n2,d2 = f2
    return simplificarFracao(criarFracao((n1*d2) + (n2*d1), (d1*d2)))
    


def subtrairFrac(f1,f2):
    n1,d1 = f1
    n2,d2 = f2
    F = criarFracao((n1*d2) - (n2*d1), (d1*d2))
    return simplificarFracao(F)


def multiplicarFrac(f1,f2):
    n1,d1 = f1
    n2,d2 = f2
    F = criarFracao(n1*n2,d1*d2)
    return simplificarFracao(F)


def dividirFrac(f1,f2):
    n1,d1 = f1
    n2,d2 = f2
    F = criarFracao(n1*d2,d1*n2)
    return simplificarFracao(F)


def criarLista():
    lista = []
    n = int(input("Introduza o número de elementos da lista: "))    
    for i in range (n):
        n = random.randrange(1,100)
        d = random.randrange(2,200)
        f = criarFracao(n,d)
        lista.append(f)
    return lista
    
    

def somarL(lista):
    soma=(0,1)
    print("Lista gerada aleatoriamente: " + str(verListFrac(lista)) )
    for elem in lista:
        soma = somarFrac(soma,elem)
    print ("A soma de todas as frações da lista é: " + str(verFracao(simplificarFracao(soma))))
    


def multiplicarL(lista):
    multi = lista[0]
    print("Lista gerada aleatoriamente: " + str(verListFrac(lista)))
    for i in range (1,len(lista)):
        multi = multiplicarFrac(multi,lista[i])
    print ("A multiplicação de todas as frações da lista é: " + str(verFracao(simplificarFracao(multi))))
    
    

def maiorFrac(lista):
    maximo = lista[0]
    for i in range (1,len(lista)):
        n1,d1 = maximo
        n2,d2 = lista[i]
        if (n1*d2) < (n2*d1):
            maximo = lista[i]
    print('O maior elemento da lista é: ' ,str(maximo[0]) + '/' + str(maximo[1]))
    
    
def menorFrac(lista):
    minimo = lista[0]
    for i in range (1,len(lista)):
        n1,d1 = minimo
        n2,d2 = lista[i]
        if (n1*d2) > (n2*d1):
            minimo = lista[i]
    print('O menor elemento da lista é: ' , str(minimo[0]) + '/' + str(minimo[1]))
    
    

def menu():
    print(""" Bem Vindo!! Seleciona uma opção: 
    (1) Criar uma Fração
    (2) Simplificar Fração
    (3) Somar Frações
    (4) Subtrair Frações
    (5) Multiplicar Frações
    (6) Dividir Frações
    (7) Gerar lista de frações aleatórias 
    (8) Somar as frações da lista
    (9) Multiplicar as frações da lista
    (10)Qual a maior fração da lista
    (11)Qual a menor fração da lista
    (0) Sair """)
    
    
opcao = "1"
while opcao != "0":
    menu()
    opcao = input("Introduza uma opção: ")
    if opcao == "1":
        print("----------Criar Fração----------")
        num = int(input("Numerador: "))
        den = int(input("Denominador: "))
        f = criarFracao(num,den)
        print("Fração: ", verFracao(f))
        
        
    elif opcao == "2":
        print("----------Simplificar Fração----------")
        num = int(input("Numerador: "))
        den = int(input("Denominador: "))
        f = criarFracao(num,den)
        print("Fração simplificada: ", verFracao(simplificarFracao(f)))
        
        
    elif opcao == "3":
        print("----------Somar Frações----------")
        n1 = int(input("Numerador da 1ª fração: "))
        d1 = int(input("Denominador da 1ª fração: "))
        n2 = int(input("Numerador da 2ª fração: "))
        d2 = int(input("Denominador da 2ª fração: "))
        f1 = criarFracao(n1,d1)
        f2 = criarFracao(n2,d2)
        s = somarFrac(f1,f2)
        print("Fração 1: ", verFracao(f1))
        print("Fração 2: ", verFracao(f2))
        print("Soma das Frações: ", verFracao(simplificarFracao(s)))
        
        
    elif opcao == "4":
        print("----------Subtrair Frações----------")
        n1 = int(input("Numerador da 1ª fração: "))
        d1 = int(input("Denominador da 1ª fração: "))
        n2 = int(input("Numerador da 2ª fração: "))
        d2 = int(input("Denominador da 2ª fração: "))
        f1 = criarFracao(n1,d1)
        f2 = criarFracao(n2,d2)
        f = subtrairFrac(f1,f2)
        print("Fração 1: ", verFracao(f1))
        print("Fração 2: ", verFracao(f2))
        print("Subtração das Frações: ", verFracao(simplificarFracao(f)))
        
        
    elif opcao == "5":
        print("----------Multiplicar Frações----------")
        n1 = int(input("Numerador da 1ª fração: "))
        d1 = int(input("Denominador da 1ª fração: "))
        n2 = int(input("Numerador da 2ª fração: "))
        d2 = int(input("Denominador da 2ª fração: "))
        f1 = criarFracao(n1,d1)
        f2 = criarFracao(n2,d2)
        m = multiplicarFrac(f1,f2)
        print("Fração 1: ", verFracao(f1))
        print("Fração 2: ", verFracao(f2))
        print("Multiplicação das Frações: ", verFracao(simplificarFracao(m)))
        
        
    elif opcao == "6":
        print("----------Dividir Frações----------")
        n1 = int(input("Numerador da 1ª fração: "))
        d1 = int(input("Denominador da 1ª fração: "))
        n2 = int(input("Numerador da 2ª fração: "))
        d2 = int(input("Denominador da 2ª fração: "))
        f1 = criarFracao(n1,d1)
        f2 = criarFracao(n2,d2)
        d = dividirFrac(f1,f2)
        print("Fração 1: ", verFracao(f1))
        print("Fração 2: ", verFracao(f2))
        print("Divisão das Frações: ", verFracao(simplificarFracao(d)))
        
        
    elif opcao == "7":
        print("----------Lista de Frações Aleatórias----------")
        lista = criarLista()
        print("A lista gerada é: ", verListFrac(lista))
       
           
        
    elif opcao == "8":
        print("----------Somar Frações da Lista----------")
        somarL(criarLista())
        
        
        
    elif opcao == "9":
        print("----------Multiplicar Frações da Lista----------")
        multiplicarL(criarLista())
        
        
    elif opcao == "10":
        print("----------Qual o maior elemento da Lista----------")
        lista = criarLista()
        print("Lista gerada aleatoriamente: ", verListFrac(lista))
        maiorFrac(lista)
        
        
    elif opcao == "11":
        print("----------Qual o menor elemento da Lista----------")
        lista = criarLista()
        print("Lista gerada aleatoriamente:  ", verListFrac(lista))
        menorFrac(lista)
    elif opcao == "0":
        print("Até à próxima!!")
    else:
        print("Opção inválida!!")
        
```
