***Resumo:*** Neste trabalho foram adicionadas duas novas funções ao código do trabalho realizado anteriormente sobre operações com polinómios. Uma das funções permite a simplificação de um polinómio sendo que, o algoritmo tem como ponto de partida a ordenação dos termos consoante os seus expoentes e, posteriormente, procede à sua simplificação. A segunda função permite a visualização da representação gráfica do polinómio.<br> 

```python
import matplotlib.pyplot as plt

def criarTermo(coef, exp):
    return (coef, exp)

def criarPolinomio():
    return []

def inserirTermo(p,t):
    p.append(t)
    
    
def verTermo(t):
    c, e = t
    if e == 0:
        return str(c)
    elif c == 1:
        return "x^" + str(e)
    else:
        return str(c) + "x^" + str(e)
        

        
def verPolinomio(p):
    politexto = ""
    if len(p) != 0:
        politexto = verTermo(p[0])
        for i in range(1,len(p)):
            if p[i][0] >= 0:
                politexto = politexto + "+" + verTermo(p[i])
            else:
                politexto = politexto + verTermo(p[i])
    return politexto



def grauPol(p):
    grau = p[0][1]
    for i in range(1,len(p)):
        if p[i][1] > grau:
            grau = p[i][1]
    
    return grau




def calcPol(p,x):
    valor = 0 
    for (c,e) in p:
        valor = valor + c * x**e
    return valor 



def derivadaPol (p):
    derivada = []
    for (c, e) in p:
        if e != 0:
            derivada.append((c*e, e-1))
    return derivada 



def tabela(p, linhas):
    print("x" + " :: " + verPolinomio(p))
    for l in range(0, linhas):
        print(str(l) + " :: " + str(calcPol(p, l)))
        

        
def simplificaSort(p):
    
    ret = []
    ordenar(p)
    i = 0
    while i < len(p):
        j = i+1
        soma = p[i][0]
        print('Elem: ' + str(p[i]))
        while j < len(p) and p[i][1] == p[j][1]:
            soma += p[j][0]
            j = j+1
        ret.append((soma,p[i][1]))
        i = j      
    return ret


def ordenar(p) :
    
    qSort(p,0,len(p)-1)
    p.reverse()
    return p


def qSort(p, low, high):    
    if low < high:
        pi = partition(p, low, high)
        qSort(p, low, pi-1)  
        qSort(p, pi+1, high)
        
        
def partition (arr, low, high):
    pivot = arr[high] 
    i = (low - 1)  
    for j in range (low,high+1):
        if arr[j][1] < pivot[1] :
            i = i+1   
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i+1





def grafico(p):
    Xinf = int(input("Limite inferior das abcissas: "))
    Xsup = int(input("Limite inferior das abcissas: "))
    if Xinf > Xsup:
        print("Inseriu os valores pela ordem errada!")
    x = [n for n in range(Xinf, Xsup+1)]
    y=[]
    for n in x:
        y.append(calcPol(p,n))
    plt.plot(x, y, color="purple", linewidth=2, marker="o", markerfacecolor="green", markersize=4)
    eixoXx=[x[0]-1,x[-1]+1]
    eixoXy=[0,0]
    plt.plot(eixoXx, eixoXy, color="green", linewidth=1)
    supy=calcPol(p,Xsup)
    infy=calcPol(p,Xinf)
    if (supy > infy):
        eixoYy=[infy-1,supy+1]
    else:
        eixoYy=[supy-1,infy+1]
    eixoYx=[0,0]
    plt.plot(eixoYx, eixoYy, color="green", linewidth=1)
    plt.xlabel("Abcissas")
    plt.ylabel("Ordenadas")
    plt.title("Gráfico do polinómio:")
    plt.show()




def menu():
    print("""Bem vindo!! Seleciona uma opção: 
    (1) Criar um polinómio
    (2) Listar o polinómio
    (3) Grau do polinómio
    (4) Calcular o polinómio num ponto
    (5) Calcular a derivada do polinómio 
    (6) Simplificar o polinómio
    (7) Tabela
    (8) Representação Gráfica
    (0) Sair """)

opcao = "1"
while opcao != "0":
    menu()
    opcao = input("Introduza uma opção: ")
    if opcao == "1":
        print("----------Criar Polinómio----------")
        p = criarPolinomio()
        n = int(input("Quantos termos terá o polinómio?"))
        for i in range(n):
            coef = int(input("Introduza o coeficiente do termo " + str(i+1) + ":"))
            exp = int(input("Introduza o expoente do termo " + str(i+1) + ":"))
            inserirTermo(p,(coef,exp))
        criarPolinomio()
    elif len(p) == 0:
            print("O polinómio ainda não foi criado!")
            
    elif opcao == "2":
        print("----------Listar Polinómio----------")
        print(verPolinomio(p))
    
    elif opcao == "3":
        print("----------Grau do Polinómio----------")
        print("O grau do polinómio é: ", grauPol(p))
    elif opcao == "4":
        print("----------Calcular Polinómio num Ponto----------")
        x = int(input("Insira o valor para o qual quer calcular o polinómio: "))
        print("O valor do polinómio em " + str(x) + "é: ", calcPol(p,x))
    elif opcao == "5":
        print ("----------Calcular a Derivada do Polinómio----------")
        print("A derivada do polinómio é: ", verPolinomio(derivadaPol(p)))
    elif opcao == "6":
        print("----------Simplificar o Polinómio----------")
        print("Polinómio simplificado: ", verPolinomio(simplificaSort(p)))
    elif opcao == "7":
        print("----------Tabela----------")
        nlinhas = int(input("Número de linhas da tabela: "))
        print("x :: Valor")
        tabela(p,nlinhas)
    elif opcao == "8":
        print("----------Representação Gráfica----------")
        grafico(p)
    elif opcao == "0":
        print("Até à próxima!!")
    else:
        print("Opção inválida!!")
```
